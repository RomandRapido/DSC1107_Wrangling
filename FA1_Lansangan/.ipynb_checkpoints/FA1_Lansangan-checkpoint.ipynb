### Tidy Data
For a data to be considered as Tidy:
1. Each variable has its own column.
2. Each observation has its own row.
3. Each value must have its own cell.

We'll explore this further on the following concepts.

### Pivoting
If the data isn't tidy, pandas have fairly easy tools to transform data into tidy data. 

Let's first start by creating non-tidy dataset. 

The values of the data will not make sense but let us focus on their structure rather than the logical consistency of the numbers in them.

```python
date = ["1999", "2000"]
data_points = 1000
district_id = []
for i in range(int(data_points/2)):
    district_id.append(i)
    district_id.append(i)

date_val_1 = [random.randint(100, 20000) for x in range(data_points)]
date_val_2 = [random.randint(100, 20000) for x in range(data_points)]
type_val = ["cases" if x%2==1 else "population" for x in range(data_points)]

non_tidy = pd.DataFrame({"district_id" : district_id, "type": type_val, '1999': date_val_1, '2000': date_val_2})
print("Shape: ", non_tidy.shape)
non_tidy.head()
```

#### a) Longer Pivot
This is needed if data have columns that should be values for a new variable. In our artificial data set, there are columns named "1999" amd "2000" which can be values for a new column `year`.

```python
tidy_longer = pd.melt(
    non_tidy,
    id_vars=['district_id', 'type'],
    value_vars=['1999', '2000'],
    var_name='year',
    value_name='count'
)
print("Shape: ", tidy_longer.shape)
tidy_longer.head()
```

Notice the the number of observations increased from 1000 to 2000 (from their shapes), hence the name 'longer'.

```python
print(tidy_longer.groupby(['district_id', 'type', 'year']).size().reset_index(name="count"))
```

#### b) Wider Pivot
This is needed if a value in a coloumn should be separated in their own columns. In our case, `population` and `cases` values of column `type` could have their own columns. 

```python
tidy = tidy_longer.pivot(
    index=['district_id', 'year'],
    columns='type',
    values='count'
).reset_index()
tidy.columns.name = ""
tidy.head()
```

Notice the number of columns increased from 3 to 4, hence the name "wider."

Our dataset is now tidy!

### Joining
Most of the time, the columns needed for our objectives came from various differing table, that's where joining comes in. This technique is used to join two table based on equivalent column from each table.

For example, here's a table of coordinates for each district

```python
n_districts = int(data_points/2)
district_id = [x for x in range(n_districts)]
X = [random.randint(-200000, 20000) for x in range(n_districts)]
Y = [random.randint(-200000, 20000) for x in range(n_districts)]
lat_lon = [f'{x},{y}' for x,y in zip(X,Y)]
coordinates = pd.DataFrame({"district_id": district_id, 'lat-lon': lat_lon})
coordinates.head()
```

We shall now *join* the tidy table to the coordinates table on the basis of their `district_id`.

```python
joined = tidy.merge(coordinates, on='district_id')
joined.head()
```

### Separating
Sometimes, separation of values is needed for atomicity (it isn't very tidy to not be one). In the case of the table above, we opt to separate `lat-lon` column into two separate columns `lat` and `lon`. Noticed that we could use the character `,` to separate the two values.

```python
joined[['lat','lon']] = joined['lat-lon'].str.split(',', expand=True).astype(int)
# drop the redundant column
joined.drop(columns='lat-lon', inplace=True)
joined.head()
```

### Missing Values
It is important to look for missing values (or NaN) because they may lead to unforeseen complications and errogenous conclusions. As the old adage goes, *"Garbage in, garbage out"*.

There plenty of ways to count for Null or Missing Values, as well as how to handle them.

#### a) Loooking for missing values

```python
joined.loc[1, 'lat'] = np.nan 
joined.loc[3, 'lon'] = np.nan 
```

```python
joined.info()
```

You can compare the number of columns and the `Non-Null Count` for each columns. If the Non-Null Count is lesser than the number of rows, there are empty values in their corresponding columns.

Another simple method is the following:

```python
joined.isna().sum()
```

#### b) Handling Missing Values
Dependending on the use case and the number of observations there are plenty of ways to handle. 


##### Drop Missing Values
One is to just simply delete the rows with missing values

```python
joined.dropna().isna().sum()
```

##### Average per Column / Imputation
Another way is to take the average of the column with missing values and set the missing value to that derrived average. 
This will work for any Central Tendency, **mean** and **median** for continuous variables---**mode** for both continuous and categorical variables.

```python
average = joined.select_dtypes("number").median()
joined_imputed = joined.fillna(average)
print("Before imputation\n", joined.head(4))
print("\n\nAfter imputation\n", joined_imputed.head(4))
```

### Isolating Data

#### a) Filter/Subset Data

Sometimes, we only want to analyse a portion of the data, this is called sampling in statistics. 
To do so in pandas and python, we can create a mask with all necessary conditions.

Say we want to look for diamonds of at least 3 carats and less than $10,000.

```python
mask = (df['carat'] > 3) & (df['price'] < 10000)
mask.head()
```

Notice that we gotten a series with boolean values. We can then use this as a mask to our original dataframe so that it will only retain the rows that has *True* value.

```python
df[mask]
```

It appears that there are 3 diamonds that meets our criteria!! We getting rich!?

#### b) Select Columns
We can also split the data horizontally by only including the columns we want.

```python
# Columns to show
cols = ['carat', 'cut', 'price']
select = df[cols]
select.head()
```

#### c) Arrange/Sort
We can also sort the dataframe by their values on a specific column or more.

```python
# Sort the data by 'carat' in descending order and price by ascending 'order'
select.sort_values(by=['carat', 'price'], ascending=[False, True]).head()
# use 'select' table derrived above for easier interpretability. But it will work on the original dataframe
```

### Derriving Information

#### a) Mutate
If we want to add a column as function of existing columns, say we want to find the deviation of a `price` of a diamond from other the average price of all diamonds,

```python
avg_price = df['price'].mean()
df['price_dev'] = df['price'] - avg_price
df.head()
```

Pandas could also handle more complex task like computing for a 'good' value diamonds.

```python
# Good value if [price < 5000 and carat > 2] or [carate <= 2 and price < 1000]
cond = (
    ((df['price'] < 5000) &
    (df['carat'] > 2)) |
    ((df['price'] < 1000) &
    (df['carat'] <= 2))
) 
df['good_value'] = cond

# Good color if color is either D, E, F, G
cond = (df['color'].isin(['E', 'D', 'F', 'G']))
df['good_color'] = cond
df.head()
```

#### b) Summarise
Used to calculate summary statistics.

### Multi-step transformations

#### a) Group By
Group by is mainly used for grouping observations of same categorical variable together.

For example, if we want to find the maximum price and depth of all diamonds by clarity, 

```python
df.groupby('clarity')[['price', 'depth']].max()
```

We can take this up further by grouping the observations based on the combination of two or more variables.

```python
df.groupby(['clarity', 'color', 'cut'])['price'].max().to_frame()[1:10]
```

#### b) Pipe (in R) / Method Chainging (in Python)
We can continously transorm the data by *method chaining*.

```python
## Average Price of Diamonds of At least one carats
df[(df['carat']>= 1)]['price'].mean()
```

```python
(df.apply(lambda x: x['price']/ x['depth'],axis=1)
.groupby(df['cut'])
.mean()
.rename("Avg Price per Depth")
.reset_index())
```

#### c) Storing the transformed data
Applying transformation techniques doesn't change the original data by default.

```python
df.head()
```

If we want to save, we'll have to equate the transformation methods to a variable name.

```python
avg_price_per_depth = (
    df.apply(lambda x: x['price']/ x['depth'],axis=1)
    .groupby(df['cut'])
    .mean()
    .rename("Avg Price per Depth")
    .reset_index())
avg_price_per_depth
```

## DATA VISUALIZATION

### Variation

#### a) Discrete Variables
What is the most common kind of diamond cut?

```python
sns.set_theme(style='darkgrid')

count_clarity = df['cut'].value_counts(normalize=True).sort_values()
sns.barplot(x=count_clarity.index, y=count_clarity.values, palette='BuPu', hue=count_clarity.index, legend=False)
plt.xlabel('Cut')
plt.ylabel('Proportion')
plt.title('Clarity Distribution')
plt.grid(linestyle="dashed", alpha=0.9);
```

The most common diamond cut is the *Ideal* diamond cut with around 40% of the total diamonds in the dataset. *Fair* on the other hand is the least common type of diamond cut, making up less than 5% of the total sample space.  

#### b) Continuous Variables
What can be said about the variation in carat?

```python
sns.histplot(df['carat'], legend=False, bins=20)
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.title('Carats Distribution')
plt.grid(linestyle="dashed");
```

Carats distribution appears to be skewed to the right with most diamonds being less than 1 carats.

### Covariation

#### a) Discrete versus discrete
What can we say about the relationship between cut and clarity?

```python
count_per_carats = df.pivot_table(index='cut', columns='clarity', aggfunc='size', fill_value=0)
count_per_carats
```

```python
melted = count_per_carats.reset_index().melt(id_vars='cut', var_name='clarity', value_name='count')

sns.scatterplot(data=melted, x='cut', y='clarity', size='count', sizes=(10, 200), hue='count', palette="rocket_r")

plt.title("Scatter Plot of Counts by Cut and Clarity")
plt.xlabel("Cut")
plt.ylabel("Clarity")
plt.legend(title="Count", bbox_to_anchor=(1, 1))
plt.show()
```

As was indicated in the bar plot of distribution of `cut` earlier, significant amount of diamonds are under this category regardless of their `clarity`. Nonetheless, it is clear to see that most diamonds are **ideal cut and VS2 clarity**. 

#### b) Continuous versus continuous

##### Many to One
In here, we shall use `x` (length in mm) and `y` (width in mm) columns due to the fact thata single `x` could be associated by many `y`.

```python
sns.scatterplot(data=df, x='x', y='y')

plt.title("Length vs Width Scatter Plot")
plt.xlabel("Length [mm]")
plt.ylabel("Width [mm]")
plt.show()
```

Let's remove the outliers by cutting down the upper and lower 5% for each variable.

```python
x_1, x_2 = df['x'].quantile([0.05, 0.95])
y_1, y_2 = df['y'].quantile([0.05, 0.95])

mask = (df['x'].between(x_1, x_2)) & (df['y'].between(y_1, y_2))

sns.scatterplot(data=df[mask], x='x', y='y')

plt.title("Length vs Width Scatter Plot, Without Outliers")
plt.xlabel("Length [mm]")
plt.ylabel("Width [mm]")
plt.show()
```

It is clear to see that there is a strong correlation between length (`x`) and width (`y`) of diamonds. Lets test this out by plotting y=x.

```python
sns.scatterplot(data=df[mask], x='x', y='y', label="Data Points")

y_1, y_2 = x_1, x_2
plt.plot([x_1, x_2], [y_1, y_2], color='red', label="y = x", linestyle='--', linewidth=3)

plt.title("Length vs Width Scatter Plot, Without Outliers")
plt.xlabel("Length [mm]")
plt.ylabel("Width [mm]")
plt.legend()
plt.show()
```

##### One to One
In here, let's find the average `price` for each `carat`.

### Additional Visualization tools

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['carat'], legend=False, bins=20)
plt.xlabel('Carat')
plt.ylabel('Frequency')
plt.title('Carats Distribution')
plt.grid(linestyle="dashed");

plt.subplot(1, 2, 2)
sns.histplot(np.log(df['carat']), legend=False, bins=20)
plt.xlabel('Log(Carat)')
plt.ylabel('Frequency')
plt.title('Carats Distribution, Log Transformed')
plt.grid(linestyle="dashed");
```
