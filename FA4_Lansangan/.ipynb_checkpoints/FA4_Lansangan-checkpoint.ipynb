### 1.4	Model (12 points)
There are clearly some trends in this data, but they are somewhat hard to see given the substantial amount of variability. This is where splines come in handy.

#### Split
To ensure unbiased assessment of predictive models, letâ€™s split the data before we start modeling it.

1.	Split bmd into training (80%) and test (20%) sets, using the rows in train_samples below for training. Store these in tibbles called bmd_train and bmd_test, respectively.

```python
bmd_train, bmd_test = train_test_split(
    df, test_size=0.2, random_state=42
)

print(f"Shape of bmd_train:", bmd_train.shape)
print(f"Shape of bmd_test:", bmd_test.shape)
```

#### 1.4.2	Tune (10 points)

Since the trendsin spnbmd look somewhat different for boys than for girls, we might want to fit separate splines to these two groups. Separate bmd_train into bmd_train_male and bmd_train_female, and likewise for bmd_test.

```python
bmd_train.sort_values(by='age', inplace=True)
bmd_train_male = bmd_train[bmd_train['sex'] == "M"]
bmd_train_female = bmd_train[bmd_train['sex'] == "F"]
```

2.	Using cross_validate_spline from the stat471 R package, perform 10-fold cross-validation on bmd_train_male and bmd_train_female, trying degrees of freedom 1,2,. . . ,15. Display the two resulting CV plots side by side.

```python
def ns(x, df=None, knots=None, boundary_knots=None, include_intercept=False):
    degree = 3
    if boundary_knots is None:
        boundary_knots = [np.min(x), np.max(x)]

    if df is not None:
        nIknots = df - 1 - include_intercept
        if nIknots < 0:
            nIknots = 0
            
        if nIknots > 0:
            knots = np.linspace(0, 1, num=nIknots + 2)[1:-1]
            knots = np.quantile(x, knots)
        else:
            knots = np.array([])

    Aknots = np.sort(np.concatenate((boundary_knots * 4, knots)))
    n_bases = len(Aknots) - (degree + 1)
    
    basis = np.empty((x.shape[0], n_bases), dtype=float)
    for i in range(n_bases):
        coefs = np.zeros((n_bases,))
        coefs[i] = 1
        basis[:, i] = splev(x, (Aknots, coefs, degree))

    const = np.empty((2, n_bases), dtype=float)
    for i in range(n_bases):
        coefs = np.zeros((n_bases,))
        coefs[i] = 1
        const[:, i] = splev(boundary_knots, (Aknots, coefs, degree), der=2)

    if include_intercept is False:
        basis = basis[:, 1:]
        const = const[:, 1:]

    qr_const = np.linalg.qr(const.T, mode='complete')[0]
    basis = (qr_const.T @ basis.T).T[:, 2:]

    return basis, knots
```

```python
class StatsModelsOLS(BaseEstimator, RegressorMixin):
    def __init__(self, formula):
        self.formula = formula
        self.model_ = None

    def fit(self, X, y=None):
        data = X.copy()
        data['target'] = y
        self.model_ = smf.ols(formula=self.formula, data=data).fit()
        return self

    def predict(self, X):
        return self.model_.predict(X)

def add_natural_spline(df, column, df_value, include_intercept=False):
    basis, knots = ns(df[column].values, df=df_value, include_intercept=include_intercept)
    for i in range(basis.shape[1]):
        df.loc[:, f'ns_{i}'] = basis[:, i] 
    return df, knots

cv = 10
result = []
model_t = []

for n in tqdm(range(1, 16)):
    if n < 3:
        formula = f"target ~ {' + '.join([f'np.power(age,{i+1})' for i in range(n)])}"
    else:
        bmd_train_male, knots_male = add_natural_spline(bmd_train_male, 'age', df_value=n, include_intercept=False)
        bmd_train_female, knots_female = add_natural_spline(bmd_train_female, 'age', df_value=n, include_intercept=False)
        formula = f"target ~ {' + '.join([f'ns_{i}' for i in range(n)])}"

    model = StatsModelsOLS(formula=formula)
    
    scores_mal = -cross_val_score(model, bmd_train_male, bmd_train_male['spnbmd'], cv=cv, scoring='neg_root_mean_squared_error')
    mean_mal, std_mal = scores_mal.mean(), scores_mal.std()
    
    scores_fem = -cross_val_score(model, bmd_train_female, bmd_train_female['spnbmd'], cv=cv, scoring='neg_root_mean_squared_error')
    mean_fem, std_fem = scores_fem.mean(), scores_fem.std()

    result.append({"n": n, "sex": "male", "mean": mean_mal, "std": std_mal})
    result.append({"n": n, "sex": "female", "mean": mean_fem, "std": std_fem})
    model_t.append(model.fit(X=bmd_train_female, y=bmd_train_female['spnbmd']))
```

```python
plt.figure(figsize=(15, 6))

plt.scatter(bmd_train_female['age'], bmd_train_female['spnbmd'], 
            color='gray', alpha=0.5, label='Data (female)')

ages = np.linspace(bmd_train_female['age'].min(), 
                   bmd_train_female['age'].max(), 100)
df_plot = pd.DataFrame({'age': ages})

upto = 15
degree_of_freedoms = range(1, upto+1) 

for n, fitted_model in zip(degree_of_freedoms, model_t[0:upto]):
    if n >= 3:
        df_plot,knots = add_natural_spline(df_plot, 'age', df_value=n, include_intercept=False)
    
    y_pred = fitted_model.predict(df_plot)
    
    plt.plot(ages, y_pred, label=f'n={n}', alpha=0.8)

plt.xlabel('Age')
plt.ylabel('spnbmd')
plt.title('Models of Different Degrees of Freedom (Female)')
plt.legend()
plt.show()
```

Looking at the graph, it appears as the best degrees of freedom to minimise *RMSE* is `df=1`. But to be perfectly candid, we can't go wrong here because all through 1-10, we have relatively low MSE! Which is amazing.

The best degrees of freedom, `df=1`, for both males and females, aren't splining method because a natural cubic spline requires at least 3 degrees of freedom. The workaround I did is to conduct a polynomical regression for $df  < 3$. So `df=1`, as it turns out, is just a linear regression and is a better model than cubic splining. This is a sign that further model complexity isn't always ideal. Nonetheless, let us proceed by only considering those that are $df \geq 3$.

```python
male_data = male_data[male_data['n'] >= 3]
female_data = female_data[female_data['n'] >= 3]
```

3.		What are the degrees of freedom values minimizing the CV curve for boys and girls, and what are the values obtained from the one standard error rule?

The **one-standard-error** is picking the best degrees of freedom by looking for the lowest RMSE score among the cross-validation result and adding it's corresponding standard error/deviation to the RMSE. The result of such operation then becomes a threshold for picking the lowest degrees of freedom among those that have lower RMSE than the result.

```python
min_rmse_index = male_data['mean'].idxmin()
min_rmse = male_data.loc[min_rmse_index]
threshold_male = min_rmse['mean'] + min_rmse['std']
print("Threshold for male:", threshold_male)
```

```python
min_rmse_index = female_data['mean'].idxmin()
min_rmse = female_data.loc[min_rmse_index]
threshold_female = min_rmse['mean'] + min_rmse['std']
print("Threshold for female:", threshold_female)
```

```python
top_mals = male_data[male_data['mean'] < threshold_male].sort_values(by='mean').head()
top_mals
```

```python
top_fems =female_data[female_data['mean'] < threshold_female].sort_values(by='mean').head()
top_fems
```

4.		For the sake of simplicity, letâ€™s use the same degrees of freedom for males as well as females. Define df.min to be the maximum of the two df.min values for males and females, and define df.1se likewise. Add these two spline fits to the scatter plot of spnbmd (y axis) versus age (x axis), faceting by gender.

```python
df_min_mal = top_mals.sort_values(by="mean").iloc[0]['n']
df_min_fem = top_fems.sort_values(by="mean").iloc[0]['n']

df_1se_mal = top_mals.sort_values(by="n").iloc[0]['n']
df_1se_fem = top_fems.sort_values(by="n").iloc[0]['n'] 

df_min = max(df_min_mal, df_min_fem)
df_1se = max(df_1se_mal, df_1se_fem)

print("Df_min:", df_min)
print("Df_1se:", df_1se)
```

It appears that the lowest `df=3` gives us the lowest RMSE. Which is furthers reinforced the idea that more complex model doesn't necessarily equate to better result.

Because of that, to compensate, let's just create a model for each `sex` with the chosen degrees of freedom `df=3`.

```python
n = 3
bmd_train_male, knots_male = add_natural_spline(bmd_train_male, 'age', df_value=n, include_intercept=False)
formula = f"target ~ {' + '.join([f'ns_{i}' for i in range(n)])}"
model_male = StatsModelsOLS(formula=formula)
model_male.fit(X=bmd_train_male, y=bmd_train_male['spnbmd'])
```

```python
bmd_train_female, knots_female = add_natural_spline(bmd_train_female, 'age', df_value=n, include_intercept=False)

formula = f"target ~ {' + '.join([f'ns_{i}' for i in range(n)])}"
model_female = StatsModelsOLS(formula=formula)
model_female.fit(X=bmd_train_female, y=bmd_train_female['spnbmd'])
```

```python
bmd_df, knots = add_natural_spline(df, 'age', df_value=n, include_intercept=False)

df_plot = bmd_df.copy()

df_plot['spnbmd_pred'] = np.nan

male_mask = df_plot['sex'] == "M"
df_plot.loc[male_mask, 'spnbmd_pred'] = model_male.predict(df_plot[male_mask])

female_mask = df_plot['sex'] == "F"
df_plot.loc[female_mask, 'spnbmd_pred'] = model_female.predict(df_plot[female_mask])

df_plot_male = df_plot[df_plot['sex'] == "M"]
df_plot_female = df_plot[df_plot['sex'] == "F"]
```

2.	How do the training and test errors compare? What does this suggest about the extent of overfitting that has occurred?

The difference between `train_rmse` and `test_rmse` from both `sex` suggests that we have indeed chose right degrees of freedom with 3 because it have prevented overfitting. 2.	The splines help us see the trend in the data much more clearly. Eyeballing these fitted curves, answer the following questions. At what ages (approximately) do boys and girls reach the peaks of their growth spurts? At what ages does growth largely level off for boys and girls? Do these seem in the right ballpark?

The fitted curves indicate that males reach their highest `spnbmd` values, or the peak of their growth spurts, in their mid-to-late 40s to early 50s, with growth largely leveling off in their 70s. In contrast, females appear to peak earlier, around their early-to-mid 40s, followed by a gradual decline through their 50s and 60s, with growth stabilizing by about 70. These trends align well with clinical observations, where women often experience a significant decline in bone density around menopause (typically in the late 40s to early 50s), while men maintain peak density slightly longer before a gradual decrease in later years (Finkelstein et al., 2007).
